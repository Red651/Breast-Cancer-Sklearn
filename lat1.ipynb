```python
import os
import random
import cv2
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
SIZE = 256
image_dataset = []
mask_dataset = []
```

```python
image_directory = 'data'
mask_directory = 'label'
```

```python
num_data = 320
images = os.listdir(image_directory)
images.sort()
images = images[0:num_data]
for i, image_name in enumerate(images):
    if (image_name.split(sep='.')[1] == 'png'):
      image = cv2.imread(image_directory+ '/'+ image_name,0)
      image = Image.fromarray(image)
      image = image.resize((SIZE,SIZE))
      image_dataset.append(np.array(image))
num_images = len(images)
```

```python
masks = os.listdir(mask_directory)
masks.sort()
masks = masks[0:num_data]
for i, image_name in enumerate(masks):
    if (image_name.split(sep='.')[1] == 'png'):
      image = cv2.imread(mask_directory+ '/'+ image_name,0)
      image = Image.fromarray(image)
      image = image.resize((SIZE,SIZE))
      mask_dataset.append(np.array(image))
print("Total number of  images are:", num_images)
```

```python
image_dataset = np.expand_dims(image_dataset, axis = 3)
mask_dataset = np.expand_dims(mask_dataset, axis = 3)
```

```python
image_dataset = image_dataset /255.
mask_dataset = mask_dataset /255.
```

```python
seed = 47
np.random.seed = seed
img_num = random.randint(0, num_images)
img_for_plot = cv2.imread(image_directory+'/'+images[img_num],0)
mask_for_plot = cv2.imread(mask_directory+'/'+masks[img_num],0)
plt.figure(figsize=(12,8))
plt.subplot(121)
plt.imshow(img_for_plot, cmap='Purples')
plt.title('Image')
plt.subplot(122)
plt.imshow(mask_for_plot,cmap='gray')
plt.title('Mask')
plt.show()
```
